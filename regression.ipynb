```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from model import MultipleRegression, normalize
```

```python
df = pd.read_csv('advertising_and_sales_clean.csv')
df = df.drop(columns=['influencer'])

df = normalize(df)
```

```python
model  = MultipleRegression(learning_rate = 0.1, batch = 150, epochs = 10)
train_data, test_data  = model.test_split(df[['tv', 'radio', 'sales']], 0.15)

model.train(train_data[['tv', 'radio']], train_data['sales'], cost_funtion=True)

print(model.get_accuracy(test_data[['tv', 'radio']], test_data['sales']))
```

```python
# %matplotlib widget

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

X, Y = np.meshgrid(x, y)
Z = (model.weights[0, 0] * X) + (model.weights[1, 0] * Y) + model.bias

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')

ax.scatter(df['tv'][:300], df['radio'][:300], df['sales'][:300], label="Actual Dataset" , color='red', s=7)
ax.plot_surface(X, Y, Z, label="Trained Model", alpha=0.5, shade=False)

ax.set_title('Model Plane')
ax.set_xlabel('Feature-1')
ax.set_ylabel('Feature-2')
ax.set_zlabel('Label')
ax.legend()
```
