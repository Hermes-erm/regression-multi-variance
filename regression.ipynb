```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from model import MultipleRegression, normalize
```

```python
df = pd.read_csv('advertising_and_sales_clean.csv')
df = df.drop(columns=['influencer'])
# print(df.head()) # 1000

# %matplotlib widget

df = normalize(df) # standardize -> check on later

# fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
# ax.scatter(df['tv'], df['radio'], df['sales'] ,s=5)
```

```python
model  = MultipleRegression(learning_rate = 0.1, batch = 300, epochs = 10)
train_data, test_data  = model.test_split(df[['tv', 'radio', 'sales']], 0.15)

model.train(train_data[['tv', 'radio']], train_data['sales'], cost_funtion=True)

# print(*model.loss)
print(model.get_accuracy(test_data[['tv', 'radio']], test_data['sales']))
```
